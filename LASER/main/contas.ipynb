```python
c = 299792458
```

```python
def FSR_teorico(L, Erro_L):
    fsr = c/(2*L)
    error = np.abs(-c/(2*L*L))*Erro_L
    return fsr, error
```

```python
print(FSR_teorico(0.0725, 0.001))
print(FSR_teorico(0.0611, 0.001))
```
